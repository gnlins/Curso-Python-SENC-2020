##Lógica de programação

Podemos definir “lógica de programação” como a elaboração de sequências de ações para atingir um determinado objetivo. O processo envolve o uso de dispositivos lógicos, como estruturas condicionais (if/else) e de repetição (for/while). ## Algoritmo

Algoritmo é uma sequência de passos ordenados para chegar em um objetivo.

## Python


Python é uma linguagem de programação de alto nível, orientada a objetos. Foi lançada por Guido van Rossum em 1991. A linguagem foi projetada com a filosofia de enfatizar a importância do esforço do programador sobre o esforço computacional. Prioriza a legibilidade do código sobre a velocidade ou expressividade. Combina uma sintaxe concisa e clara com os recursos poderosos de sua biblioteca padrão e por módulos e frameworks desenvolvidos por terceiros. ## Variáveis


As variáveis armazenam objetos nos nossos códigos. Assim como na matemática temos que 'x' pode valer 10, 20, 30, ou outros infinitos valores, na programação as variáveis podem assumir valores numéricos, textos, coleções, objetos, comandos, ...

As variáveis são definidas (ou declaradas) da seguinte forma:

**variável = valor**

1. Primeiro o nome da variável. Ao escolher o nome é importante se atentar a algumas limitações: A variável não pode começar com número, não pode ser alguma palavra reservada e não pode conter espaços ([Recomendação de leitura](https://aprendacompy.readthedocs.io/pt/latest/capitulo_02.html))
2. O sinal de igualdade colocado **apenas uma vez** representa atribuição de valor
3. O valor que aquela variável irá assumir


Exemplos:

a = 10

b = "Esse é o melhor curso de Python que já fiz"

c = 3 + 5


nome = "Gustavo"


**IMPORTANTE:** A ordem dos fatores altera o resultado, então antes da igualdade sempre será a variável e depois o valor que ela irá assumir.

```
#Declarando uma variável
a= 10
b= "Esse é o melhor curso de Python que já fiz"
c = 3+5
```

##Operadores Matemáticos

São frequentes as situações que precisamos fazer operações matemáticas por meio da programação, portanto é importante saber quais são os operadores que o Python usa para cada caso. Os operadores são:

1. Soma: +
2. Subtração: -
3. Multiplicação: *
4. Divisão: /
5. Potenciação: **
6. Divisão inteira: //
7. Resto da divisão inteira: %

**Exemplos:**

3 + 4 = 7

5 - 2 = 3

5 * 2 = 10

10 / 2 = 5

5 ** 2 = 25

7//2 = 3

7%2 = 1

```
op1 = 3 + 4 

op2 = 5 - 2 

op3 =5 * 2 

op4 =10 / 2 

op5 =5 ** 2 

op6 =7//2 

op7 =7%2 
```

## 4. Faça um programa que receba do usuário as informações Nome, peso e altura e que retorne um texto informando o IMC (Índice de massa corporal) da seguinte forma: "Olá _**Nome**_, sua altura é _**altura**_ e seu peso é _**peso**_, então seu IMC é: _**IMC**_"

```
nome = input("Diga seu nome")
altura = float(input("Qual é sua altura, em metros?"))
peso = float(input("Qual é o seu peso?"))

IMC = peso/altura**2

print("Olá {}, sua altura é {} seu peso é {}, então seu IMC é: {}".format(nome,altura,peso,IMC))
```
