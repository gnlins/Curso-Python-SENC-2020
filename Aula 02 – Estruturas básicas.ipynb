## Operadores Relacionais ou comparativos

Operador Relacional é todo operador que obtém o resultado da relação do membro a sua esquerda com o membro a sua direita.

>Descrição | Operador
>--- | ---
>Maior que | >
>Menor que | <
>Igual a | ==
> Diferente | !=
>Maior ou igual a| >=
>Menor ou igual a | <=
> Contido em uma sequencia|**in**




````
a = 10
b = 5
c = 5
d = 'Python'

print(a>b)
print(a<b)
print(b>=c)
print(b<=c)
print(b==c)
print(b!=c)
print('p' in d)
````

## Operadores Lógicos

Os operadores lógicos unem expressões lógicas formando assim, uma nova expressão que é composta por 2 ou mais sub-expressões.  

>Descrição | Operador
>--- | ---
>ou | or
>e | and
>não | not

````
print(a>b and c<b)
print(a>b or c<b)
print(not a>b)
````

#Estrutura Condicionais

As estruturas condicionais são utilizadas para a tomada de decisão, ou seja, se algumas linhas do código irão ser executadas ou não.

Condicional Simples: 

![Estrutura Condicional](http://fabrica.ms.senac.br/wp-content/uploads/2013/06/imagem2.jpg)

Condicional composta:

![Condicional composta](http://fabrica.ms.senac.br/wp-content/uploads/2013/06/imagem3-300x194.jpg)

Estruturas condicionais do Python:
- **if** : O if é uma estrutura de condição que permite avaliar uma expressão e, de acordo com seu resultado, executar uma determinada ação

- **else** : Caso a condição avaliada pelo if não seja atendida, definimos o fluxo alternativo que o código deve seguir

- **elif**: Se existir mais de uma condição alternativa que precisa ser verificada, devemos utilizar o elif para avaliar as expressões intermediárias antes de usar o else

````
a = 7
b = 3

if a>b:                     #<-Aqui está a minha primeira condição
  print("a é maior que b")  #<-bloco de ação que será executado, caso a primeira condição seja verdadeira
elif a<b:                   #<-Segunda condição
  print("a é menor que b")  #<-bloco de ação que será executado, caso a primeira condição seja falsa e a segunda verdadeira
else:                       #<-indica o que será feito nas demais condições
  print('a é igual a b')    #<-Bloco que será executado se nenhuma condição anterior for verdadeira
````

````
nome = input("Coloque seu nome")
if nome == "Gustavo":
  print("Olá, Gustavo")
else:
  print("Olá {}, acho que não te conheço rsrs".format(nome))
````


#Estruturas de Repetição

Estruturas de repetição são utilizadas quando desejamos executar uma parte do código várias vezes obedecendo algum critério de parada.

## **for**

Use o for quando você tem um número predeterminado de repetições a se realizar. Foco no intervalo

![Fluxograma for](https://image.slidesharecdn.com/apresentao-170510192311/95/captulo-2-python-62-638.jpg?cb=1494444228)

````
for elemento in range(1,6): #Para o elemento dentro do intervalo estabelecido, vou fazer o que estiver no bloco de ação
  print(elemento)           #Bloco de ação, identado após os dois pontos, indicando tudo que será repetido
````

````
l1 = ['a','b','c','d', 'e']
string = "Olá, meu nome  é Gustavo!"
for indice,elemento in enumerate(string):
  print(indice, elemento)
````

## **while**

Quando a intenção não é predefinir quantas vezes uma determinada tarefa deverá ser executada. Foco na condição.

![Fluxograma while](https://image.slidesharecdn.com/a09lm31213javascriptrepeticoes-121022064057-phpapp02/95/a09-labmm3-javascript-estruturas-de-repetio-14-638.jpg?cb=1350888140)

````
numero = int(input("Tente adivinhar o numero"))
while numero !=10: #Inicio da estrutura de repetição condicional, junto com a condição que possibilita a repetição
  numero = int(input("Tente novamente")) #O que deverá ser repetido, no caso a solicitação de um novo número para o usuário
print("Parabéns, vocÊ acertou")
````

````
for e in range(1,11):
  print(e)
````

````
n = 1
while n <=10:
  print(n)
  n = n+1
````

# **Funções**

Na programação, funções são blocos de código que realizam determinadas tarefas que normalmente precisam ser executadas diversas vezes dentro de uma aplicação. Quando surge essa necessidade, para que várias instruções não precisem ser repetidas, elas são agrupadas em uma função, à qual é dado um nome e que poderá ser chamada/executada em diferentes partes do programa. Fonte: [DevMedia](https://www.devmedia.com.br/funcoes-em-python/37340)

````
def f(x, z):
  y = 5*x+7*z
  return y

print(f(1,3))
f(2,0)
````

````
def adivinhe(numero):
  chute = int(input("Qual é o numero"))
  while chute != numero:
    chute = int(input("Qual é o numero"))
  print("PArabéns, vocÊ acertou!")
````

````
adivinhe(5)
````
