#Ciência de Dados

## O que é Ciência de dados?

Ciência de Dados é a responsável por trazer insights. Isso é, dado uma grande quantidade de dados, que conclusões podemos tirar a partir deles? 

Ela é um campo interdisciplinar que utiliza métodos, processos, algoritmos e sistemas científicos para extrair valor dos dados. Os cientistas de dados combinam uma série de habilidades, incluindo estatísticas, ciência da computação e conhecimento comercial, para analisar dados coletados da web, smartphones, clientes, sensores e outras fontes.

De acordo com a IDC, empresa de inteligência de mercado e consultoria em serviços estratégicos, geramos 2,5 quintilhões de bytes de dados todos os dias.

Fontes: [Caleum](https://blog.caelum.com.br/qual-e-a-diferenca-de-data-science-machine-learning-e-inteligencia-artificial/), [Oracle](https://www.oracle.com/br/data-science/what-is-data-science.html) e [Techagr](https://techagr.com/o-que-os-nossos-dados-podem-nos-ensinar-entenda-o-que-o-big-data-muda-na-pecuaria/#:~:text=Entendendo%20o%20conceito,de%20dados%20todos%20os%20dias.)

## O que é Inteligência Artificial?

<center><img src="https://levee.com.br/wp-content/uploads/2019/08/video-POST-MACHINE-LEARNING.jpg" alt="Método do trapézio" width="700" /></center>

Dois termos que são frequentemente utilizados são, Machine Learning e Inteligência artificial, mas o que são essas duas coisas?

Segundo Michal Colins, no Livro _Machine Learning, an intruduction to Supervised & Unsupervised Learning Algorithms_, Machine Learning é sobre dar sentido ao grande volume de dados, tendo em vista que quando trabalhamos com uma enorme quantidade de dados, dificilmente conseguimos analisar de forma crítica.

De forma simplificada, _Machine Learning_ é a forma com que os computadores aprendem ao serem apresentados aos dados.

Quando a máquina é apresentada aos dados, ela aprende e consegue apresentar diversos resultados: Um curva que melhor se ajusta aos pontos, reconhecer uma figura, classificar um e-mail se é spam ou não. A Inteligência artificial é a ação que o computador executa após a interpretação daquele dado

## Numpy

O NumPy é uma biblioteca Python que é usada principalmente para realizar cálculos em Arrays Multidimensionais. O NumPy fornece um grande conjunto de funções e operações de biblioteca que ajudam os programadores a executar facilmente cálculos numéricos.

Fonte: [Medium](https://medium.com/ensina-ai/entendendo-a-biblioteca-numpy-4858fde63355)

Abaixo seguem alguns dos seus comandos principais

```
import numpy as np #Importando a biblioteca numpy, chamando-a de np
```

```
#Criando uma matriz
n = [[5,7,3],[10,8,5],[1,2,3]]
M = np.array(n).reshape((3,3))
O = np.arange(5)

print(M)
print(O)
```

```
#MAtriz de zeros
zeros = np.zeros((5,5))
zeros
```

```
#Matriz de uns
ones = np.ones((5,5))
ones
```

```
#Matriz Identidade
Identidade = np.eye(4)
Identidade
```

```
#Shape
Identidade.shape
```

```
#Reshape
Identidade.reshape((2,8))
```

```
a = np.arange(1,7).reshape((2,3))
b = np.arange(1,10).reshape((3,3))
c = np.eye(3)

print(a,'\n')
print(b,'\n')
print(c,'\n')
```

```
#operações de matriz
print(b+c,'\n')
print(b-c,'\n')
print(b*c,'\n')
print(c/b,'\n')
```

```
#Multiplicação de matriz
a.dot(c)
```

```
#Determinante
determ = np.linalg.det(n)
print(determ)
```

```
#Tranposta
M.T
```

```
#Inversa
Minv = np.linalg.inv(M)
```

```
Minv.dot(M)
```
